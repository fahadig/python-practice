In pandas, the dt.strftime() method is used to format datetime objects as strings. This method allows you to specify the format of the date and time using specific format codes. Each format code represents a different component of the date-time object, such as year, month, day, hour, etc.

Here's a detailed breakdown of commonly used format symbols in pandas with dt.strftime():

Commonly Used Format Codes
%Y: Year with century as a decimal number. Example: "2024".
%y: Year without century as a zero-padded decimal number. Example: "24".
%m: Month as a zero-padded decimal number. Example: "04" for April.
%B: Month as locale’s full name. Example: "April".
%b: Month as locale’s abbreviated name. Example: "Apr".
%d: Day of the month as a zero-padded decimal number. Example: "01".
%A: Weekday as locale’s full name. Example: "Sunday".
%a: Weekday as locale’s abbreviated name. Example: "Sun".
%H: Hour (24-hour clock) as a zero-padded decimal number. Example: "14".
%I: Hour (12-hour clock) as a zero-padded decimal number. Example: "02".
%p: Locale’s equivalent of either AM or PM. Example: "AM".
%M: Minute as a zero-padded decimal number. Example: "05".
%S: Second as a zero-padded decimal number. Example: "09".
%f: Microsecond as a decimal number, zero-padded on the left. Example: "000999".
%z: UTC offset in the form +HHMM or -HHMM. Example: "+0000".
%Z: Time zone name. Example: "UTC".
%j: Day of the year as a zero-padded decimal number. Example: "095".
%W: Week number of the year (Monday as the first day of the week) as a decimal number. Example: "13".
%U: Week number of the year (Sunday as the first day of the week) as a decimal number. Example: "14".
%c: Locale’s appropriate date and time representation.
%x: Locale’s appropriate date representation.
%X: Locale’s appropriate time representation.
Example Usage
To use dt.strftime(), you first need a pandas Series with datetime type. Here’s how you can format a date column:

python
Copy code
import pandas as pd

# Sample data

# Format the 'date' column

This will create a new column formatted_date with the dates formatted as "YYYY-MM-DD".

Summary
The strftime() function in pandas is highly flexible and allows you to display datetime objects in any format you need by using the appropriate format codes. This can be especially useful when preparing data for reports or converting it for user-friendly displays.
