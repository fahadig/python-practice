```python
from typing import TextIO

data:TextIO = open("./abc.txt") #connectivity with abc.txt file
print(data.read()) #reads all data in one go, which is dangerous for memory

data.close() #close connectivity
```

```python
with open("./abc.txt") as file: #TextIO --> WITH block closes the connectivity outside the block automatically
    print(type(file))
    print(file.read())

print("Pakistan")
```

```python
with open("./abc.txt") as file: 
    print(type(file))
    print(file.readline()) #reads ONLY first line at a time
    print(file.readline()) 
    
    
```

```python
with open("./abc.txt") as file: 
    print(type(file))
    print(file.readline(),end="")
    print(file.readline(),end="") # how to remove \n by putting end
```

```python
with open("./abc.txt") as file: #WITH block close connectivity outside the block automatically
    print(type(file))
    print(file.readlines()) #reads with iteration type
```

```python
with open("./abc1.txt", 'w') as file: # w has overwrite all previous data to Pakistan Zindabad
   
    
    file.write("Pakistan Zindabad")
```

```python
with open("./abc1.txt", 'r+') as file: # w has overwrite all previous data to Pakistan Zindabad
    print(file.read()) # before read and write
        
    file.write("We love our country")
    file.seek(0)

    print(file.read()) # after read and write

```
