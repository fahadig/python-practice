# multi input and multi output

```python
def fn_obtained(s1:int, s2:int, s3:int, s4:int, s5:int)->tuple[int, int, float, str]:
    total : int = 500
    obtained : int = s1 + s2 + s3 + s4 + s5
    per : float = obtained / total * 100
    grade : str = ""
    if per >= 90:
        grade ='A+'
    elif per >= 80:
        grade = 'A'
    elif per >= 70:
        grade = 'B'
    elif per >= 60:
        grade = 'C'
    elif per >= 50:
        grade = 'D'
    elif per >= 33:
        grade = 'E'
    else:
        grade = 'Fail'
    
    return total,obtained,per,grade

df1[['Total','Obtained','Percentage','grade']] = df1[["s1","s2","s3","s4","s5"]].apply(lambda x:fn_obtained(*x),axis=1,result_type='expand') # Single Input and multi Output
df1 #lamba(*x) = each value of each subject from df tupple and will add it in the variable, and you will get the answer
```
